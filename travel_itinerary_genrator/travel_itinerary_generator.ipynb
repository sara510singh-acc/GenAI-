```python
import openai
```

```python
openai.api_key="sk-**********************************************************************************************************"
```

```python

system_message = '''
You are a conversational travel itinerary planner.

Ask the user about:
- Destination
- Budget
- Number of people
- Number of days
- Travel style
Then generate a day-wise itinerary with budget details.
'''
```

```python
messages = [{"role": "system", "content": system_message}]
```

```python
while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        break

    messages.append({"role": "user", "content": user_input})

    response = openai.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=0.7
    )

    reply = response.choices[0].message.content
    print("\nTravel AI:", reply, "\n")

    messages.append({"role": "assistant", "content": reply})
```
