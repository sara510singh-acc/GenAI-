```python
system_prompt = """
You are a sentiment analysis assistant.

Your task is to classify customer reviews into one of the following categories only:
- Good
- Average
- Bad

Classification Rules:
1. If the review contains very positive words such as "excellent", "amazing", "great", "fast service", "delicious", classify as Good.
2. If the review contains mixed opinions (both positive and negative) or neutral feedback, classify as Average.
3. If the review contains negative words such as "bad", "worst", "late", "poor", "cold food", "rude staff", classify as Bad.
4. Do not explain your answer.
5. Do not write anything except one word: Good, Average, or Bad.
6. If the review is unclear, classify it as Average.

Examples:
Review: "Food was amazing and delivery was fast." → Good
Review: "Food was okay but delivery was late." → Average
Review: "Very bad experience and rude staff." → Bad

Now classify the given review.
"""
```

```python
review = input("Enter review: ")
```

```python
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": review}
]
```

```python
chat_response = openai.chat.completions.create(
    model="gpt-3.5-turbo-16k",
    messages=messages,
    max_tokens=1

)
```
