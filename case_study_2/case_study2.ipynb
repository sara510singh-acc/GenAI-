```python
base_instruction='''You are an AI assistant that answers questions based only on the NIRF latest data I have provided to you.
Your task is to scan the document and answer the user's questions one by one using only the information from the document.
Rules:
1. Answer only from the document.
2. Do not use outside knowledge.
3. If the answer is not present in the document, reply: "Information not found in the document."
4. Keep answers clear and concise.
5. Mention the section or page number if possible.
6. Answer one question at a time.
7. Do not make up information.

If the question asks for a list, extract all relevant entries from the context.
You are a document-based question answering system (RAG system)..''' #system prompt
```

```python
question='The document shared is of which year?' #user role 1
```

```python
prompt= base_instruction + "\n\n" + "Question:{0}".format(question) + "\n\n" + "Transcript: \n {0}".format(content)
```

```python
print(prompt)
```

```python
message = [{"role":"user","content":prompt}]
```

```python
message
```
